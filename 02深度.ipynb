```python
#今日练习：数据操作+自动求导
import torch
x= torch.tensor([[1.0,2.0],[3.0,4.0]])
print("x 的形状：x.shape")
print("x:\n",x)
      
```

```python
import torch
x = torch.tensor([1.0,2.0,3.0],requires_grad = True)
print("初始张量 x:",x)
```

```python
#构建计算图
y = x**2
z = y.sum()
print("计算结果z：",z)
#反向传播计算梯度
z.backward()
```

```python
#查看梯度
print("x的梯度:",x.grad)
```

```python
#向量梯度计算示例
x_vector =torch.tensor([1.0,2.0],requires_grad = True)
y_vector = x_vector*3
y_vector.backward(torch.tensor([1.0,1.0]))
print("向量x的梯度",x_vector.grad)
```

```python
#梯度清零
x.grad.zero_()
print("清零后x的梯度是:",x.grad)
```
